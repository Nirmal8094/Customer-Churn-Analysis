```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = sns.load_dataset("tips")
print(data)
```

```python
c = sns.color_palette("viridis")
sns.lineplot(data = data,x = "day",y = "tip",hue = "sex",style = "sex",palette = c)
plt.show()
```

```python
#Bar Plot
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
sns.barplot(data,x = "day",y = "tip" ,estimator = "mean" )
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
#sns.barplot(data,x = "day",y = "tip" ,estimator = "sum" )
#sns.barplot(data,x = "day",y = "tip" ,estimator = "median" ,errorbar = ("ci",0),hue = "sex",palette = "viridis")
sns.set_style("whitegrid")
sns.barplot(data,x = "day",y = "tip",estimator="mean",errorbar=("ci",0),hue = "sex",hue_order=["Female","Male"],palette="viridis",
            order = ["Sun","Fri","Sat","Thur"])
plt.show()
```

```python
#Hist Plot seaborn
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset("tips")
sns.histplot(data,x = "day",hue = "sex")
plt.plot()
```

```python
sns.histplot(data,x = "day",shrink = 0.9)
plt.show()
```

```python
#Scatter Plot 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
print(data)
sns.scatterplot(data,x = "total_bill",y = "tip",hue = "sex",style = "smoker",size ="size",palette = "spring" )
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset("tips")
gp = data.groupby("day").agg({"tip":"mean"})
sns.heatmap(gp)
plt.show()
```

```python
#Count Plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
#print(data)
sns.countplot(data,x = "sex",hue = "day",hue_order=["Sun","Sat","Thur","Fri"],order=["Female","Male"])
plt.show()
```

```python
#Multiple Plot
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = sns.load_dataset("tips")
#print(data)
a = sns.FacetGrid(data,col="smoker",hue = "sex",dogde = True)
a.map(sns.barplot, "day", "tip",errorbar = ("ci",0))
plt.show()
```
